1. 
Обратная строка
Напишите функцию, которая принимает строку и возвращает её в обратном порядке.
Пример: "hello" → "olleh".



```python
word=input('введите слово')
drow=word[::-1]#срез строки с шагом -1, то есть перебор символов с конца к началу
print('перевёрнутое слово;',drow)
```

```python
word=input('введите слово')
drow=''.join(reversed(word))#reversed(word) возвращает итератор с символами в обратном порядке;
#''.join(...) склеивает символы в одну строку, если в кавычках наипсать символ он будет использоваться для вставки междусимволами
print('перевёрнутое слово:',drow)
```

2.
Подсчёт гласных
Подсчитайте, сколько гласных (a, e, i, o, u) содержится в заданной строке (регистр не учитывать).
Пример: "Hello World" → 3.



```python
def count_volwes(s):
    volwes={'a','e','i','o','u'}
    count=0 #оператор начинает счёт с 0
    for char in s.lower():# .lower делает из всех прописных на вводе, строчные на выводе
        if char in volwes: #char - тип данных для хранения 1 символа
           count+=1 #шаг сдвига счётчика 
    return count
#работает по принцупу: нашёл одну букву из словаря, добавляет к счёту от 0 +1, и так 
def count_consonants(c):
    consonants = {'b', 'c', 'd', 'f', 'g', 'h', 'j', 'k', 'l', 'm', 'n', 'p', 'q', 'r', 's', 't', 'v', 'w', 'x', 'z'}
    count=0
    for char in c.lower():
        if char in consonants:
            count+=1
    return count
text=input('введите строку')
print('введённая строка:', text)
print('гласные:')
print(count_volwes(text))
print('согласные:')
print(count_consonants(text))




```
